### Objective:
Scrape product data from Nykaa using 3 category links:

Then combine all the data into one dataset for EDA.


### importing all libraries useful for webscrapping:

#### get request of makeup products

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
all_products = []

for i in range(1, 201):   # 1 to 201 pages

    url = f"https://www.nykaa.com/search/result/?q=makeup%20products&root=search&searchType=history&suggestionType=query&ssp={i}"

    r = requests.get(url)
    print("Page:", i, "Status Code:", r.status_code)

    if r.status_code != 200:
        continue

    soup = BeautifulSoup(r.text, "html.parser")

    names = soup.find_all("div", class_="css-xrzmfa")
    selling = soup.find_all("span", class_="css-111z9ua")
    mrps = soup.find_all("span", class_="css-17x46n5")
    disc = soup.find_all("span", class_="css-cjd9an")
    tags = soup.find_all("li", class_="custom-tag")
    rate=soup.find_all("span",class_="css-1qbvrhp")  # example


    for j in range(len(names)):

        product_name = names[j].text.strip()
    
        selling_price = selling[j].text.strip() if j < len(selling) else ""
        mrp = mrps[j].text.strip() if j < len(mrps) else ""
        discount = disc[j].text.strip().replace("Off", "").strip() if j < len(disc) else ""
        tag = tags[j].text.strip() if j < len(tags) else "None"
    
        rating = rate[j].text.strip() if j < len(rate) else ""
    
        all_products.append({
            "Product Name": product_name,
            "MRP": mrp,
            "Selling Price": selling_price,
            "Discount": discount,
            "Tag": tag,
            "Rating": rating
        })

# DataFrame
df1 = pd.DataFrame(all_products)


df1
```

```python
def get_brand(product_name):
    words = product_name.split()

    if len(words) == 0:
        return ""

    if len(words) == 1:
        return words[0]

    # mostly brand is 1 or 2 words
    first = words[0]
    second = words[1]

    # if second word is not brand type then keep only 1 word
    not_brand_words = ["New", "Pro", "Professional", "Matte", "Ultra", "Perfect", "Long", "Liquid", "Super"]

    if second in not_brand_words:
        return first
    else:
        return first + " " + second


df1["Brand Name"] = df1["Product Name"].apply(get_brand)

df1

```

```python
df1.insert(0, "Category", "Makeup")
```

```python
df1["Product Name"] = df1["Product Name"].apply(lambda x: re.sub(r"[^A-Za-z0-9 ]", "", x))
```

```python
df1.to_csv("nykaa_makeup.csv",index=False)
print("CSV Saved Successfully")
```
